```python
list_png=os.listdir()
for png in list_png[:]: 
    if not(png.endswith(".png")):
        list_png.remove(png)
print(list_png)
```

```python
#fetch top n color 
def FetchTopColor(picname,n):
    colorlist = []
    im = Image.open(list_png[0]) 
    # Quantize down to n colour palettised image using *"Fast Octree"* method:
    q = im.quantize(colors=n,method=2)
    for i in range(0,n*3,3):
        color = q.getpalette()[i:i+3]
        colorlist.append(color)
    return colorlist
```

```python
#dirname = os.path.dirname(__file__)
def changeColor(image_file,colorlist,x,TopColor):
    im = Image.open(image_file)
    data = np.array(im)
    c = randint(0,50)
    ###########changecolor###########
    for i in range(0,TopColor-1):
        seed(c+x+i) 
        r1, g1, b1 = colorlist[i][0], colorlist[i][1], colorlist[i][2] # Original value
        r2,g2,b2 = randint(0,256),randint(0,256),randint(0,256)
        red, green, blue = data[:,:,0], data[:,:,1], data[:,:,2]
        mask = (red == r1) & (green == g1) & (blue == b1)
        data[:,:,:3][mask] = [r2, g2, b2]
    image_file = image_file.replace(".png","")
    im = Image.fromarray(data)
    im.save(os.path.dirname(os.path.realpath('__file__')) + '/results/' + (str(x)) + '.png')
    saved=image.imread(os.path.dirname(os.path.realpath('__file__')) + '/results/' + (str(x)) + '.png')
    pyplot.imshow(saved)
    pyplot.show()

    return f"Done {image_file}"
```

```python
a=19841984 #randome seed change to whatever number you like
TopColor= 25 # fetch top N color
freq_list = []
pic_num = 50 
for i in range(0,pic_num ): # generate 
    seed(a+i)
    print('This is',str(i)+'.png')
    b=randint(0,len(list_png)-1)
    selectedimage = image.imread(list_png[b])
    colorlist =FetchTopColor(list_png[b],TopColor)
    new_image = changeColor(list_png[b],colorlist, i,TopColor)  
    freq_list.append(b)
    if i >= pic_num -1:
        print('Done!')
        pass
    else:
        pyplot.imshow(selectedimage)
```

```python
#check image frequency if you need
def CountFrequency(my_list):
    # Creating an empty dictionary 
    freq = {}
    for item in my_list:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
    for key, value in freq.items():
        print ("% d : % d"%(key, value))
CountFrequency(freq_list)
```
